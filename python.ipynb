```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 生成rgb三種顏色的10*20圖片
images = np.zeros((3, 10, 20, 3), dtype=np.uint8)
gray_images = np.zeros((3, 10, 20), dtype=np.uint8)
# 初始化統計陣列
max_value = np.zeros(3, dtype=np.uint8)
min_value = np.zeros(3, dtype=np.uint8)
mean_value = np.zeros(3, dtype=np.uint8)
std_value = np.zeros(3, dtype=np.uint8)

images[0, :, :, 0] = np.random.randint(0, 256, size=(10, 20))  # Red channel
images[1, :, :, 1] = np.random.randint(0, 256, size=(10, 20))  # Green channel
images[2, :, :, 2] = np.random.randint(0, 256, size=(10, 20))  # Blue channel

#轉換灰階
for i in range(3):
    gray_images[i,:,:] = np.mean(images[i,:,:,:], axis=2).astype(np.uint8)
    max_value[i] = np.max(gray_images[i,:,:])
    min_value[i] = np.min(gray_images[i,:,:])
    mean_value[i] = np.mean(gray_images[i,:,:])
    std_value[i] = np.std(gray_images[i,:,:])
gray_images = np.mean(images, axis=3).astype(np.uint8)
max_value[0] = np.max(gray_images[0, :, :])
fig, axs = plt.subplots(2, 3, figsize=(12, 8))
for i in range(3):
    axs[0, i % 3].imshow(images[i])
    axs[0, i % 3].set_title(f"Image {i+1}")
    axs[0, i % 3].axis('off')
for i in range(3):
    axs[1, i % 3].imshow(gray_images[i], cmap='gray')
    axs[1, i % 3].set_title(f"gray_image {i+1}")
    axs[1, i % 3].axis('off')
plt.tight_layout()
plt.show()
# 將結果存入Excel檔案
d= {
    '圖片編號': np.arange(1, 4),
    '最大值': max_value,
    '最小值': min_value,
    '平均值': mean_value,
    '標準差': std_value
}
df = pd.DataFrame(d)
df.to_excel('neaf test python.xlsx', index=False)





```
